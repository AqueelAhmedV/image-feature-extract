```python
import pandas as pd
from utils import download_images

# Load the training data
train_df = pd.read_csv('../dataset/train.csv')

# Download images
image_paths = download_images(train_df['image_link'], './images')

# Add image paths to the dataframe
train_df['image_path'] = image_paths
```

```python
import pandas as pd
train_df = pd.read_csv('../dataset/train.csv')
test_df = pd.read_csv('../dataset/test.csv')

# Get unique group IDs
train_group_ids = set(train_df['group_id'].unique())
test_group_ids = set(test_df['group_id'].unique())

# Find group IDs in test but not in train
new_group_ids = test_group_ids - train_group_ids

# Filter test dataset to include only these new group IDs
new_groups_test_df = test_df[test_df['group_id'].isin(new_group_ids)]

print(f"Number of group IDs in test but not in train: {len(new_group_ids)}")
print(f"Number of samples in test with new group IDs: {len(new_groups_test_df)}")
```

```python
import torch
from torchvision import models, transforms
from PIL import Image

# Load a pre-trained ResNet model
model = models.resnet50(pretrained=True)
model = torch.nn.Sequential(*list(model.children())[:-1])  # Remove the last fully connected layer
model.eval()

# Define image transformations
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

def extract_features(image_path):
    image = Image.open(image_path).convert('RGB')
    image = transform(image).unsqueeze(0)
    with torch.no_grad():
        features = model(image)
    return features.squeeze().numpy()

# Extract features for all images
train_df['features'] = train_df['image_path'].apply(extract_features)
```
